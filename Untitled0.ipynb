TWO SUM Question
leetcode.com/problems/two-sum/

```python
numbers = list(map(int, input("Enter numbers separated by spaces: ").split()))
target = int(input("Enter target: "))
result = []

for i in range(0, len(numbers)):
    for j in range(i, len(numbers)):
        if numbers[i] + numbers[j] == target:
            result = [i, j]

print(result)
```

```python
numbers = list(map(int, input("Enter numbers separated by spaces: ").split()))
target = int(input("Enter target: "))

def two_sum_hash_map(nums, target):
    num_map = {}
    for index, num in enumerate(nums):
        complement = target - num
        if complement in num_map:
            return [num_map[complement], index]
        num_map[num] = index
    return []

result_optimized = two_sum_hash_map(numbers, target)
print(f"Optimized result: {result_optimized}")
```

### Explanation of the Hash Map (Dictionary) Solution:

This solution uses a dictionary (hash map) to store numbers and their indices as we iterate through the list. This allows for very efficient lookups.

1.  **`num_map = {}`**: An empty dictionary `num_map` is initialized. This will store `(number: index)` pairs.
2.  **`for index, num in enumerate(nums):`**: We iterate through the `numbers` list, getting both the `index` and the `num` itself.
3.  **`complement = target - num`**: For each `num`, we calculate the `complement` needed to reach the `target`.
4.  **`if complement in num_map:`**: We check if this `complement` already exists as a key in our `num_map`. If it does, it means we've previously encountered a number that, when added to the current `num`, equals the `target`.
5.  **`return [num_map[complement], index]`**: If the `complement` is found, we return its index (which we stored in `num_map`) and the `index` of the current `num`. These are the two indices that sum to the `target`.
6.  **`num_map[num] = index`**: If the `complement` is not found, we add the current `num` and its `index` to the `num_map`. This makes it available for future lookups as a potential `complement` for subsequent numbers.
7.  **`return []`**: If the loop completes without finding a pair, an empty list is returned, indicating no such pair exists.

This method avoids nested loops, making it much faster for larger lists compared to the quadratic O(n^2) time complexity of the brute-force approach.
