```python
def plusone(digits):
    """
    :type digits: List[int]
    :rtype: List[int]
    """
    n = len(digits)
    for i in range(n-1, -1 ,-1):
        digits[i] += 1
        digits[i] %= 10
        if digits[i] != 0: # Check individual digit, not the whole list
            return digits
    return [1] + digits

# Corrected indentation and call method
input_list = list(map(int, input("Enter numbers separated by spaces: ").split()))
print(plusone(input_list))
```

```python
def plusone_alternative(digits):
    """
    :type digits: List[int]
    :rtype: List[int]
    """
    num = 0
    for digit in digits:
        num = num * 10 + digit

    num += 1

    return [int(d) for d in str(num)]

# Example usage with the alternative function
input_str = input("Enter numbers separated by spaces (for alternative solution): ")
alternative_input_list = list(map(int, input_str.split()))
print(plusone_alternative(alternative_input_list))
```
