```python
def is_balanced(s: str) -> bool:
    stack = []
    pairs = {")": "(", "}": "{", "]": "["}
    for char in s:
        if char in pairs.values():
            stack.append(char)
        elif char in pairs.keys():
            if not stack or stack.pop() != pairs[char]:
                return False
    return not stack # True if stack is empty

parentheses = input("enter the parentheses: ")
print(is_balanced(parentheses))
```

```python
def is_balanced_alt(s: str) -> bool:
    while True:
        original_len = len(s)
        s = s.replace('()', '')
        s = s.replace('{}', '')
        s = s.replace('[]', '')
        if len(s) == original_len: # No more replacements were made
            break
    return len(s) == 0

parentheses_alt = input("Enter the parentheses for the alternative check: ")
print(is_balanced_alt(parentheses_alt))
```
