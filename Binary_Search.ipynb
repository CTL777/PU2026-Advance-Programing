```python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# Get list input from the user
list_input = input("Enter a sorted list of numbers, separated by commas (e.g., 1,3,5,7,9): ")
try:
    my_list = sorted([int(x.strip()) for x in list_input.split(',')])
except ValueError:
    print("Invalid list input. Please enter numbers separated by commas.")
    my_list = [] # Set an empty list to avoid further errors

# Get target element input from the user
target_input = input("Enter the target element to search for: ")
try:
    target_element = int(target_input)
except ValueError:
    print("Invalid target element. Please enter a number.")
    target_element = None # Set to None to indicate invalid input

if my_list and target_element is not None:
    index = binary_search(my_list, target_element)
    print(index)
```

```python
def recursive_binary_search(arr, target, low, high):
    if low > high:
        return -1 # Element not found

    mid = (low + high) // 2

    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        return recursive_binary_search(arr, target, mid + 1, high)
    else:
        return recursive_binary_search(arr, target, low, mid - 1)

# Get list input from the user
list_input = input("Enter a sorted list of numbers, separated by commas (e.g., 1,3,5,7,9): ")
try:
    my_list = sorted([int(x.strip()) for x in list_input.split(',')])
except ValueError:
    print("Invalid list input. Please enter numbers separated by commas.")
    my_list = []

# Get target element input from the user
target_input = input("Enter the target element to search for: ")
try:
    target_element = int(target_input)
except ValueError:
    print("Invalid target element. Please enter a number.")
    target_element = None

if my_list and target_element is not None:
    index = recursive_binary_search(my_list, target_element, 0, len(my_list) - 1)
    print(index)
```
